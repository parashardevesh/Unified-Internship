#### wcMatches Columns Manipulation
- Making year column to int datatype
- Changing data type of Datetime column
- Making changes in stage column

```python
# changing data type of year column to integer
wcMatches.Year = wcMatches.Year.astype('int')
wcMatches.Year.dtype
```

```python
# changing datatype of datetime column to datetime
wcMatches.Datetime = wcMatches.Datetime.apply(lambda x: x.replace('-', ''))
```

```python
# converting Datetime column data type to datetime
print('Previous Datatype:', wcMatches.Datetime.dtype, '\n')
wcMatches.Datetime = pd.to_datetime(wcMatches.Datetime,format="mixed")
print('Updated Datatype:', wcMatches.Datetime.dtype)
```

```python
# Stage column, we have Group 1,2,3,4,5,6 and a,b,c,d,e,f,g,h 
# let's merge all the numbers categories with the alphabets
# writing a function to replace numbers cateogry with alphabets
def stageReplace(x):
    group = ['', 'A', 'B', 'C', 'D', 'E', 'F']
    for i in range(1,7):
        if x == 'Group {}'.format(i):
            x = 'Group {}'.format(group[i])
        if x == 'Play-off for third place' or x == 'Match for third place':
            x = 'Third place'
            
    return x

# applying changes to Stage
wcMatches.Stage = wcMatches.Stage.apply(stageReplace)
wcMatches.Stage.value_counts()
```

```python
# chacking percentage of nulls in wcPlayers' columns
print("Percentage of Nulls\n")
for i in wcPlayers.columns:
    nullPercent = round(100*wcPlayers.isna().sum()[i]/wcPlayers.shape[0], 2)
    print("{}: {}".format(i, nullPercent))
```

```python
# calculating trend line
x = np.arange(len(worldcup))
y = worldcup['GoalsScored'].values
slope, intercept, r_value, p_value, std_err = linregress(x, y)
trendline = slope * x + intercept
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(12, 6)) 
sns.barplot(y = worldcup['GoalsScored'], x = worldcup.Year,errorbar=None, color='lightblue') 
plt.plot(trendline, color='red', linewidth=2, linestyle='--', label='Trendline')
plt.title('Distribution of Goals Scored in FIFA World Cups')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
plt.legend()
plt.show() 
```

# Insight
- There is an upward trend in total number of goals scored in FIFA world cup year-on-year
- Highest number of goals i.e. 171 were seen in 1998 FIFA and 2014 FIFA

```python
x = np.arange(len(worldcup))
y = worldcup['Attendance'].values
slope, intercept, r_value, p_value, std_error = linregress(x, y)
trendline = x * slope + intercept
```

## Merging World Cup players and World Cup Matches Dataframe

```python
merged = pd.merge(wcPlayers, wcMatches, left_on=['RoundID','MatchID'], right_on=['RoundID','MatchID'], how='inner')
merged.head()
```

```python
final_match = merged[merged.Stage=='Final'].groupby(['Year', "Home Team Name","Away Team Name"])[['Home Team Goals', 
                                                                                                  'Away Team Goals']].mean().sort_values(by='Year', ascending=False)
fig, ax = plt.subplots(figsize=(12, 8))
final_match.plot(kind='barh', ax=ax)
plt.title("Goals scored in Final Matches for all FIFA world cups")
plt.xlabel('Goals Scored')
plt.show() 
```

# Insight
- Till 1990 no draws.
- 1994 first worldcup final draw
- 2006 2nd worldcup final draw. 
- Its hard to make who won the match, let's create a new column that only has winners or draws in that. 

```python
def winner(x):
    if x['Home Team Goals'] > x['Away Team Goals']:
        return x['Home Team Name']
    elif x['Home Team Goals'] < x["Away Team Goals"]:
        return x['Away Team Name']
    else:
        return "Draw"
```

```python
merged['Winner'] = merged.apply(winner, axis=1)
```

```python
plt.rcParams['figure.figsize'] = (14,9)
merged[merged.Stage=='Final'].groupby(['Year', 'Winner'])[['Home Team Goals', 'Away Team Goals']].mean().sort_values(by='Year', ascending=False).plot(kind='barh')
plt.title('Goals Scored By Winning Teams in each FIFA')
plt.show()
```

## Insight
- Most goals in FIFA history in finals were scored by Brazil i.e. 5 goals to win the match. 

### Summary of Analysis

1. **Goals Scored Over Time**:
   - **Increasing Trend**: The total number of goals scored in the FIFA World Cup has shown a consistent upward trend over the years. Notably, the highest number of goals (171) was recorded in both the 1998 and 2014 tournaments.
   - **Average Goals Per Match**: Despite the increase in total goals, the average goals per match have significantly decreased over time, from approximately 4-5 goals in earlier years to under 3 goals per match by 2014.

2. **Audience Attendance**:
   - **Growing Popularity**: Attendance at World Cup matches has also been on the rise. The 1994 World Cup in the USA saw the highest participation, with over 3.5 million attendees.
   - **Correlation with Goals**: Both attendance and the number of goals scored have shown an upward trend, indicating increasing excitement and engagement in the tournament over the years.

3. **Evolution of Matches**:
   - **Number of Matches**: Initially, fewer matches (less than 20) were played in each tournament, but this number began to increase significantly after 1954. From 1982 onwards, the number of matches stabilized at around 52, and then increased to 64 matches per tournament starting in 1998, remaining consistent up to 2014.

4. **Top Performing Teams**:
   - **Brazil's Dominance**: Brazil has consistently been a top contender, finishing in the top 4 a record 11 times from 1930 to 2014. Germany and Italy also frequently appear among the top teams.
   - **Home Advantage**: Teams tend to perform better when playing in their home country, highlighting the significance of the home advantage in winning matches.

5. **Scoring Patterns**:
   - **Second Half Scoring**: Both home and away teams are more likely to score the opening goal in the second half of the match.
   - **High Scoring Teams**: Turkey stands out with the highest average goal score per match (5 goals), followed by Hungary with an average of 4 goals per match.
   - **Home vs. Away Performance**: Most countries tend to score more when playing as the home team. However, some teams like Czech Republic and UAE have better average scores as the away team.

6. **Player and Team Participation**:
   - **Most Match Appearances**: Ronaldo and Klose are among the players with the highest number of match appearances in World Cup history.
   - **Top Participating Countries**: Brazil leads in player participation, followed by Italy, reflecting their deep involvement and success in the tournament.
   - **Coaching Legends**: Solari Luiz Felipe (Brazil) and Schoen Heimut (Federal Republic of Germany) have the most appearances as coaches in World Cup matches.

7. **Unique Insights**:
   - **Jersey Numbers**: The most common jersey numbers worn by players in FIFA are 1, 10, 15, 19, 14, 16, 11, 9, and 12. The number 0 signifies unknown shirt numbers and is not a valid jersey number.
   - **Final Match Trends**: Historically, there were no draws in the World Cup finals until 1994. Since then, the 1994 and 2006 finals ended in a draw, highlighting the increasing competitiveness in the final matches.
   - **Top Scoring in Finals**: Brazil holds the record for the most goals scored in a single World Cup final match, with 5 goals to secure victory.
