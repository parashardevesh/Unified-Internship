# Data Cleaning

- World Cup dataframe (worldcup) is clean. No nulls or duplicates. No anomalies. 
- World Cup Matches (wcMatches) Dataframe and World Cup Players (wcPlayers) Dataframe has nulls and duplicates. 

### Cleaning World Cup Matches Dataframe

-  all the columns after a certain point is showing nulls 
-  some data importing issue
-  in reality, there are only 4572 - 3720 = 852 rows in wcMatches df.
-  let's drop all the rows in which all the columns have nulls

```python
wcMatches.dropna(axis = 0, how = 'all', inplace= True)
wcMatches.shape
```

```python
# let's have a look at datatype for each column
wcMatches.info()
```

```python
# let's check the value_counts for each column

for i in wcMatches.columns:
    print(wcMatches[i].value_counts().head(20), '\n\n')
```

#### wcMatches Columns Manipulation
- Making year column to int datatype
- Changing data type of Datetime column
- Making changes in stage column

```python
# changing data type of year column to integer
wcMatches.Year = wcMatches.Year.astype('int')
wcMatches.Year.dtype
```

```python
# changing datatype of datetime column to datetime
wcMatches.Datetime = wcMatches.Datetime.apply(lambda x: x.replace('-', ''))
```

```python
# converting Datetime column data type to datetime
print('Previous Datatype:', wcMatches.Datetime.dtype, '\n')
wcMatches.Datetime = pd.to_datetime(wcMatches.Datetime,format="mixed")
print('Updated Datatype:', wcMatches.Datetime.dtype)
```

```python
# Stage column, we have Group 1,2,3,4,5,6 and a,b,c,d,e,f,g,h 
# let's merge all the numbers categories with the alphabets
# writing a function to replace numbers cateogry with alphabets
def stageReplace(x):
    group = ['', 'A', 'B', 'C', 'D', 'E', 'F']
    for i in range(1,7):
        if x == 'Group {}'.format(i):
            x = 'Group {}'.format(group[i])
        if x == 'Play-off for third place' or x == 'Match for third place':
            x = 'Third place'
            
    return x

# applying changes to Stage
wcMatches.Stage = wcMatches.Stage.apply(stageReplace)
wcMatches.Stage.value_counts()
```

#### Cleaning World Cup Players Dataframe

```python
# chacking percentage of nulls in wcPlayers' columns
print("Percentage of Nulls\n")
for i in wcPlayers.columns:
    nullPercent = round(100*wcPlayers.isna().sum()[i]/wcPlayers.shape[0], 2)
    print("{}: {}".format(i, nullPercent))
```

```python
# dropping columns with high percantage of null values
wcPlayers.drop(columns=['Position', 'Event'], inplace=True)
```

```python
# checking datatypes of each column
wcPlayers.info()
```

```python
# let's check value counts of each column
for i in wcPlayers.columns:
    print(wcPlayers[i].value_counts().head(20), '\n')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(worldcup[['GoalsScored', 'MatchesPlayed', 'Attendance']])
worldcup['pca1'] = principal_components[:, 0]
worldcup['pca2'] = principal_components[:, 1]
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=worldcup, x='pca1', y='pca2', hue='Year', palette='viridis', legend=False)
plt.title('PCA of World Cups Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show() 
```

```python
merged = pd.merge(wcPlayers, wcMatches, left_on=['RoundID','MatchID'], right_on=['RoundID','MatchID'], how='inner')
merged.head()
```

```python
final_match = merged[merged.Stage=='Final'].groupby(['Year', "Home Team Name","Away Team Name"])[['Home Team Goals', 'Away Team Goals']].mean()

fig, ax = plt.subplots(figsize=(12, 8))
final_match.plot(kind='barh', ax=ax)
plt.title("Final Matches Scores For all FIFA world cups")
plt.xlabel('Goals Scored')
plt.show() 
```

# Insight
- Till 1990 no draws.
- 1994 first worldcup final draw
- 2006 2nd worldcup final draw. 

```python
def winner(x):
    if x['Home Team Goals'] > x['Away Team Goals']:
        return x['Home Team Name']
    elif x['Home Team Goals'] < x["Away Team Goals"]:
        return x['Away Team Name']
    else:
        return "Draw"
```

```python
merged['Winner'] = merged.apply(winner, axis=1)
```
